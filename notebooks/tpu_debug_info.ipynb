```python
# CELL: Complete TPU Environment Info (fixed for torch_xla 2.9+)

import torch
import torch_xla
import torch_xla.core.xla_model as xm
import os
import sys

print("=" * 60)
print("TPU ENVIRONMENT INFO")
print("=" * 60)

# PyTorch / XLA versions
print(f"PyTorch: {torch.__version__}")
print(f"torch_xla: {torch_xla.__version__}")
print(f"Python: {sys.version.split()[0]}")

# XLA runtime info
print(f"\nXLA runtime initialized: True (device available)")
try:
    import torch_xla.runtime as xr
    print(f"Device type: {xr.device_type()}")
except:
    print("Device type: TPU (inferred)")

# Device info
device = xm.xla_device()
print(f"\nDevice: {device}")
print(f"Device HW: {xm.xla_device_hw(device)}")

# Check PJRT device
print(f"\nPJRT_DEVICE env: {os.environ.get('PJRT_DEVICE', 'NOT SET')}")
print(f"XLA_USE_BF16 env: {os.environ.get('XLA_USE_BF16', 'NOT SET')}")
print(f"XRT_TPU_CONFIG env: {os.environ.get('XRT_TPU_CONFIG', 'NOT SET')}")

# Check TPU driver
try:
    import subprocess
    result = subprocess.run(['ls', '-la', '/dev/'], capture_output=True, text=True)
    tpu_devices = [l for l in result.stdout.split('\n') if 'tpu' in l.lower()]
    if tpu_devices:
        print(f"\nTPU devices in /dev/: {tpu_devices}")
except:
    pass

# Test basic tensor ops
print("\n" + "=" * 60)
print("TESTING BASIC OPS")
print("=" * 60)
x = torch.randn(100, 100, device=device)
y = torch.randn(100, 100, device=device)
z = x @ y
import torch_xla.core.xla_model as xm
xm.mark_step()
print(f"Matrix multiply test: {z.shape} on {z.device}")
print("Basic ops working!")
```
